```python
#importar bibliotecas necessárias
import pandas as pd
uri="https://raw.githubusercontent.com/fivethirtyeight/data/master/star-wars-survey/StarWars.csv"
starwars=pd.read_csv(uri, encoding='latin-1')
starwars.head()
```

```python
# Create a dictionary to map English column names to Portuguese
column_name_mapping = {
    'RespondentID': 'ID do Respondente',
    'Have you seen any of the 6 films in the Star Wars franchise?': 'Você já viu algum dos 6 filmes da franquia Star Wars?',
    'Do you consider yourself to be a fan of the Star Wars film franchise?': 'Você se considera fã da franquia de filmes Star Wars?',
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'Quais dos seguintes filmes de Star Wars você já viu? Por favor, selecione todos que se aplicam.',
    'Unnamed: 4': 'Sem Nome 4',
    'Unnamed: 5': 'Sem Nome 5',
    'Unnamed: 6': 'Sem Nome 6',
    'Unnamed: 7': 'Sem Nome 7',
    'Unnamed: 8': 'Sem Nome 8',
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'Por favor, classifique os filmes de Star Wars em ordem de preferência, sendo 1 seu filme favorito na franquia e 6 seu filme menos favorito.',
    'Unnamed: 28': 'Sem Nome 28',
    'Which character shot first?': 'Qual personagem atirou primeiro?',
    'Are you familiar with the Expanded Universe?': 'Você está familiarizado com o Universo Expandido?',
    'Do you consider yourself to be a fan of the Expanded Universe?æ': 'Você se considera fã do Universo Expandido?',
    'Do you consider yourself to be a fan of the Star Trek franchise?': 'Você se considera fã da franquia Star Trek?',
    'Gender': 'Gênero',
    'Age': 'Idade',
    'Household Income': 'Renda Familiar',
    'Education': 'Educação',
    'Location (Census Region)': 'Localização (Região Censitária)'
}

starwars.rename(columns=column_name_mapping, inplace=True)
display(starwars.head())
```

```python
starwars.Gênero.value_counts('normalize').map('{:.1%}'.format)
```

```python
#comparação de genero com o gosto pela franquia
# Excluir a categoria 'Response' da coluna 'Gênero'
starwars_filtered = starwars[starwars['Gênero'] != 'Response']

# Calcular a contagem de cada combinação de Gênero e visualização da franquia
gender_seen_counts = starwars_filtered.groupby(['Gênero', 'Você já viu algum dos 6 filmes da franquia Star Wars?']).size()

# Calcular as proporções
gender_seen_proportions = gender_seen_counts / gender_seen_counts.sum()

# Formatar as proporções como porcentagem
display(gender_seen_proportions.map('{:.1%}'.format))
```

```python
#gráfico para mostrar quantidade de homens por renda e gosto pela franquia
starwars_filtered = starwars[starwars['Renda Familiar'] != 'Response']
starwars_filtered = starwars_filtered[starwars_filtered['Você já viu algum dos 6 filmes da franquia Star Wars?'] != 'Response']

starwars_filtered.groupby(['Renda Familiar', 'Você já viu algum dos 6 filmes da franquia Star Wars?']).size().unstack().plot(kind='bar',title='Quantidade de Homens por Renda e Gosto da Franquia',ylabel='Quantidade',xlabel='Renda Familiar')
```
